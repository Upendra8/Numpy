# DoT Product

```python
#1-D--inner product of vectors without complex conjugation
np.random.seed(1)
import pandas as pd
mat1=np.random.randint(10,size=(5,3))
mat2=np.random.randint(10,size=(5,3))
```

```python
mat1
```

```python
mat2
```

```python
np.dot(mat1,mat2.T) #but mat1 * mat2(np.multiply(a,b)) gives element wise multiplication
```

```python
sales_amount=np.random.randint(20,size=(5,3))
sales_amount
```

```python
weekly_sales=pd.DataFrame(sales_amount,index=['Mon','Tues','Wed','Thurs','Fri'],columns=['Almond Butter','Peanut Butter','Cashew Butter',])
```

```python
weekly_sales
```

```python
#Create prices array
prices=np.array([10,8,12])
```

```python
#Create butter prices DataFrame
butter_prices=pd.DataFrame(prices.reshape(1,3),index=["Price"],columns=['Almond Butter','Peanut Butter','Cashew Butter'])
```

```python
butter_prices
```

```python
total_sales=prices.dot(sales_amount.T)
```

```python
total_sales
```

```python
#Create daily_sales
weekly_sales['Total']=total_sales
```

```python
weekly_sales
```
